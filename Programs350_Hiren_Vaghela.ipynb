## 85 . Flatten a nested list (list containing sublists of arbitrary depth) into a single list.

```python
nested_ls = [1,[2,[3,4],5],6]

def flatten(lst):
    res = []
    for item in lst :
        if isinstance(item,list):
            res.extend(flatten(item))
        else:
            res.append(item)
    return res

print(flatten(nested_ls))
```

## 117 . Write a generator to yield all palindromic numbers between 1 and 1000.

```python
def palindrom():
    result = []

    for i in range(1,1001):
        temp = i
        rev = 0

        while temp > 0:
            r = temp % 10
            rev = rev * 10 + r
            temp = temp // 10

        if i == rev:
            result.append(i)
    return result


print(palindrom())
```

## 120 .Write a function to calculate the factorial of a number using a generator to yield intermediate factorial values, and use reduce() to compute the final result.

```python
from functools import reduce
def fact(n):
    if n < 0 :
        return "Factorial not defined for negative numbers"
    num = [x for x in range(1,n+1)]
    factorial = reduce(lambda x , y : x * y , num)
    return factorial

print(fact(5))
```

## 256 . Selection Sort

```python
arr = [5, 3, 8, 4, 2]

for i in range(len(arr)):
    min_index = i
    
    for j in range(i+1, len(arr)):
        if arr[j] < arr[min_index]:
            min_index = j
    
    arr[i], arr[min_index] = arr[min_index], arr[i]

print(arr)
```

## 257 . Insertion Sort

```python
arr = [5, 3, 8, 4, 2]

for i in range(1, len(arr)):
    key = arr[i]
    j = i - 1
    
    while j >= 0 and arr[j] > key:
        arr[j+1] = arr[j]
        j -= 1
    
    arr[j+1] = key

print(arr)
```

## 263 . Merge Sort (Recursive)

```python
def merge_sort(arr):
    if len(arr) > 1:
        mid = len(arr) // 2
        L = arr[:mid]
        R = arr[mid:]

        merge_sort(L)
        merge_sort(R)

        i = j = k = 0
        while i < len(L) and j < len(R):
            if L[i] < R[j]:
                arr[k] = L[i]
                i += 1
            else:
                arr[k] = R[j]
                j += 1
            k += 1

        while i < len(L):
            arr[k] = L[i]
            i += 1
            k += 1

        while j < len(R):
            arr[k] = R[j]
            j += 1
            k += 1
    return arr

data = [38, 27, 43, 3, 9, 82, 10]
print("Merge Sort:", merge_sort(data))
```

## 264 . Quick Sort (List Comprehension Method)

```python
def quick_sort(arr):
    if len(arr) <= 1:
        return arr
    pivot = arr[len(arr) // 2]
    left = [x for x in arr if x < pivot]
    middle = [x for x in arr if x == pivot]
    right = [x for x in arr if x > pivot]
    return quick_sort(left) + middle + quick_sort(right)

data = [10, 7, 8, 9, 1, 5]
print("Quick Sort:", quick_sort(data))
```

## 308. Word Frequency Counter (Dictionary & String Methods)

## 337 .Rearrange Positive & Negative Alternate

## 341 . Simple Countdown Timer

```python
import time

def countdown(seconds):
    while seconds:
        mins, secs = divmod(seconds, 60)
        timer = f'{mins:02d}:{secs:02d}'
        print(timer, end="\r")
        time.sleep(1)
        seconds -= 1
    print("Time's up!")

t = int(input("Enter time in seconds: "))
countdown(t)
```

## 344 . Linear Search

```python
def linear_search(arr, target):
    for i in range(len(arr)):
        if arr[i] == target:
            return i
    return -1

data = [10, 23, 45, 70, 11, 15]
find = int(input("Enter number to find: "))
result = linear_search(data, find)

if result != -1:
    print(f"Element found at index: {result}")
else:
    print("Element not found.")
```

## 345 . Recursive Factorial

```python
def factorial(n):
    if n == 1 or n == 0:
        return 1
    else:
        return n * factorial(n-1)

num = int(input("Enter a number: "))
print(f"Factorial of {num} is {factorial(num)}")
```

## 346 .Caesar Cipher (Basic Encryption)

```python
def caesar_cipher(text, shift):
    result = ""
    for char in text:
        if char.isalpha():
            start = ord('a') if char.islower() else ord('A')
            result += chr((ord(char) - start + shift) % 26 + start)
        else:
            result += char
    return result

msg = input("Enter a message: ")
s = int(input("Enter shift value: "))
print("Encrypted:", caesar_cipher(msg, s))
```

## 347 . Password Strength Validator

```python
import re

def check_password(password):
    if len(password) < 8:
        return "Too Short"
    if not re.search("[0-9]", password):
        return "Must contain a number"
    if not re.search("[A-Z]", password):
        return "Must contain an uppercase letter"
    return "Strong Password"

pwd = input("Set your password: ")
print(check_password(pwd))
```

## 350 . Fibonaci Series Usning Recursion

```python
def fibonacci_recursive(n):
    if n <= 1:
        return n
    else:
        return fibonacci_recursive(n - 1) + fibonacci_recursive(n - 2)

terms = int(input("Enter the number of terms: "))


print("Fibonacci sequence:")
for i in range(terms):
    print(fibonacci_recursive(i), end=" ")
```
